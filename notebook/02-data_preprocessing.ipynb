# Automated Data Preprocessing Pipeline

1. **drop_columns**
   - Remove unnecessary columns defined in configuration.

2. **rename_columns**
   - Rename raw dataset columns into standardized names.

3. **remove_invalid_responses**
   - Keep only valid responses where `responses_based_on_year == "yes"`.

4. **compute_response_time**
   - Calculate survey completion time in minutes using:
     `survey_end - survey_start`.

5. **verify_student_itc**
   - Validate `itc_student_id` using regex pattern and create
     `is_itc_student` flag.

6. **remove_duplicates**
   - Remove duplicate ITC student responses based on
     `itc_student_id`, keeping the best attempt based on
     `response_time_minutes`, then combine back with non-student responses.

7. **standardise_categoricals**
   - Normalize categorical values such as:
     - gender
     - province
     - academic_year
     - faculty
     - is_itc_student labels

8. **flag_speeder**
   - Identify suspiciously fast responses (speeders)
     based on response time thresholds.

9. **encode_ordinals**
   - Convert ordinal categorical responses into numeric values
     using predefined scales from configuration
     (e.g., AGREE5, FREQ5, LMS5).

```python
from time import sleep

import pandas as pd
from blended_learning.config.settings import settings
from blended_learning.preprocessing.clearner import DataCleaner
from blended_learning.utils.decorator import execution_time

@execution_time
def main():
    df = pd.read_excel(settings.path['input_path'])
    cleaner = DataCleaner(settings=settings, df=df)
    cleaner.run()


    # save to csv
    cleaner.df.to_csv(settings.path['output_cleaned_path'], index=False, encoding="utf-8-sig")

if __name__ == "__main__":
    main()
```
